```python
# Import necessary libraries
import numpy as np
import pandas as pd
```

```python
# Set seed for reproducibility
np.random.seed(0)
```

```python
# Generate an array of random numbers
data = np.random.randint(1, 501, size=(10,50))
data
```

```python
# Calculate the sum of each row
row_sums = data.sum(axis=1)
row_sums
```

```python
# Calculate the sum of each column
column_sums = data.sum(axis=0)
column_sums
```

```python
# Calculate the sum of all entries
data_sum = data.sum()
data_sum
```

```python
# Generate descriptive statistics for the array
print('Minimum: ', data.min())
print('Maximum: ', data.max())
print('Average: ', data.mean())
print('Standard Deviation: ', data.std())
```

```python
# To sort using pandas, I have to turn the array into a dataframe
df = pd.DataFrame(data)
df.head()
```

```python
# Sort the dataframe on row 2
row_sort = df.sort_values(by=2, axis=1)
row_sort.head()
```

```python
# Sort the dataframe on column 2
column_sort = df.sort_values(by=2)
column_sort.head()
```
